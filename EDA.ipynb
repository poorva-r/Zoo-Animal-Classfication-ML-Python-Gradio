```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("class2.csv")
df.head(20)
```

```python
plt.plot(df["ClassNo"], df["NoOfSpecies"], color = 'b',marker='o', linestyle='--', linewidth = 2, markersize = 5)
plt.title("Species in Each Class", fontsize = 12)
plt.xlabel("Class Number")
plt.ylabel("Number of Species")
plt.legend(["Species Line"],loc=4)
plt.show()
```

```python
from matplotlib import style
style.use("ggplot")
plt.plot(df["ClassNo"], df["NoOfSpecies"],"bo--",linewidth=2,markersize=9)
plt.grid(color="k",linestyle='-',linewidth=1) #k means black color
plt.show()
```

```python
df2 = pd.read_csv("zoo1.csv")
df2.head()
```

```python
plt.figure(figsize = (16,9))
plt.hist(df2["legs"], rwidth=0.9,color="m")
plt.title("Histogram")
plt.xlabel("Number of legs")
plt.ylabel("Number of animals")
plt.show()
```

```python
x=["Mammal", "Bird","Fish", "Invertebrate", "Bug", "Reptile","Amphibian" ]
y = df2['class_type'].value_counts().tolist()
```
